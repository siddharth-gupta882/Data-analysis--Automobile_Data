# Most expensive car company name

```python
df = pd.read_csv('Automobile_data.csv')
```

```python
df.nlargest(1,"price")
```

```python
high=df['price'].max()
high
```

```python
df.loc[df['price']==high,['company','price']]
```

# Details of all Toyota cars

```python
df.loc[df['company'] == 'toyota']
```

# Total cars per company

```python
df.groupby('company').count()['index']
```

# The car of each company which costs highest among other cars

```python
df.groupby('company')['price'].max()
```

# The average mileage providing car company name

```python
df.groupby('company')['average-mileage'].max()
```
